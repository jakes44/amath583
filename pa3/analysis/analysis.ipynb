```python
import matplotlib.pyplot as plt
import csv
```

```python
x = []
y = []

with open('../programs/daxpy.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
    
    for row in plots:
        x.append(int(row[0]))
        y.append(float(row[1]))

plt.plot(x, y, color = 'r', label = "Time Taken for N Operations")
plt.xlabel('N (Size of Matrix)')
plt.ylabel('FLOPS')
plt.title('N vs FLOPS for DAXPY Operation')
plt.legend()
plt.grid()
plt.show()
```
